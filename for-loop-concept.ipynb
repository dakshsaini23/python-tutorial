```python
# table of a given number
n=int(input("enter the number"))
for i in range(1,11):
    print(n,"*",i,"=",n*i )
```

```python
# factorial of a given number
n=int(input("enter the number"))
f=1
for i in range(1,n+1):
    f=f*i
print("the factorial is",f)
```
